```python

import pandas as pd
import numpy as np

#사용자가 픽한 정보
Ex = pd.DataFrame({'성별':[1],
                   '컬러':[7],
                   '톤':[4],
                   '패턴':[10],
                   '상하의':[1],
                   'TPO,스타일,무드':['일상,데이트,학교,미팅,파티남자미니멀,시크'] })

train_data = pd.read_csv('sample_data/combine.csv')
TPO = list(train_data['TPO,스타일,무드'])
y = list(train_data['class'])
Exx = list(Ex['TPO,스타일,무드'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack, csr_matrix
#TF-IDF, 
vectorizer = TfidfVectorizer(min_df=0.0, analyzer='char', sublinear_tf=True, ngram_range=(1,3), max_features=5000)
X = vectorizer.fit_transform(Exx)
Xtrain = hstack([X, csr_matrix(Ex[['컬러','톤','패턴']])])
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

RANDOM_SEED = 42
TEST_SPLIT = 0.2

y = np.array(y)

X_train, X_eval, y_train, y_eval = train_test_split(Xtrain, y, test_size=TEST_SPLIT)
```

```python
from sklearn.linear_model import LogisticRegression

lgs = LogisticRegression(class_weight = 'balanced')
lgs.fit(X_train, y_train)
```

```python
print("Accuracy: {}".format(lgs.score(X_eval, y_eval)))
```
